공연건수

```python
x = ['1월', '2월', '3월', '4월', '5월', '6월', '7월']
y1 = last_perfo['공연건수'][1:-1]
y2 = this_perfo['공연건수'][1:-1]

plt.figure(figsize=(10,6))
plt.title('공연예술 통계: 공연건수', size=20)
plt.xlabel('월')
plt.ylabel('건수(건)')

plt.plot(x, y1, label='2019년', marker='o')
plt.plot(x, y2, label='2020년', marker='o')
plt.legend()

plt.show()
```
